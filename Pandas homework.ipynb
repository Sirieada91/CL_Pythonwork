```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(12,6))
social_media_df['YouTube'].plot.hist(ax=axes[0], bins=50, alpha=0.5, color='orange')
axes[0].set_title('YouTube')

social_media_df['Twitter'].plot.hist(ax=axes[1], bins=50, alpha=0.5, color='Black')
axes[1].set_title('Twitter')

social_media_df['LinkedIn'].plot.hist(ax=axes[2], bins=50, alpha=0.5, color='green')
axes[2].set_title('LinkeDin')
```
